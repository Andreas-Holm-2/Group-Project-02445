```python
# Assuming normal distribution
n = 100
z = 1.96
labels = ["GPT 3.5","GPT 4o","LLama 70B","Gemini","Claude 3 Sonnet"]
accuracies = [[0.69,0.99,0.66,0.92,0.95],[0.7,0.93,0.76,0.62,0.90],[0.73,0.84,0.86,0.85,0.87],[0.56,0.9,0.7,0.7,0.78]]

theme = ["Commonsense","justice","Virtue and ethics", "Deontological"]

for k,j in enumerate(theme): 
    print(f"--------{j}---------")
    conf_intervals = []
    for i,p in enumerate(accuracies[k]): 
        sd = sqrt(p*(1-p)/n)
        conf = [p-z*sd, p+z*sd]
        conff = [round(c, 3) for c in conf]
        print(f"General accuracy of {labels[i]}", conff)
        conf_intervals.append(conf)

    conf_intervals = np.array(conf_intervals)
    lower_bounds = conf_intervals[:, 0]
    upper_bounds = conf_intervals[:, 1]
    errors = [accuracies[k] - lower_bounds, upper_bounds - accuracies[k]]

    x = np.arange(len(labels))
    width = 0.5

    fig, ax = plt.subplots()
    bars = ax.bar(x, accuracies[k], width, yerr=errors, capsize=5, color='lightblue')

    # Add labels, title, and custom x-axis tick labels
    ax.set_xlabel('Models')
    ax.set_ylabel('Accuracy')
    ax.set_title(f'Accuracies with Confidence Intervals on {j}')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.set_ylim(0, 1.05)  # Setting the limit of y-axis to a little above 1 for better visibility

    # Display the plot
    #plt.show()



print("------general-------")
n = 400
accuracies = [0.67, 0.915, 0.745, 0.7725, 0.875]
conf_intervals = []
for i,p in enumerate(accuracies): 
    sd = sqrt(p*(1-p)/n)
    conf = [p-z*sd, p+z*sd]
    conff = [round(c, 3) for c in conf]
    print(f"General accuracy of {labels[i]}", conff)
    conf_intervals.append(conf)

conf_intervals = np.array(conf_intervals)
lower_bounds = conf_intervals[:, 0]
upper_bounds = conf_intervals[:, 1]
errors = [accuracies - lower_bounds, upper_bounds - accuracies]

x = np.arange(len(labels))
width = 0.5

fig, ax = plt.subplots()
bars = ax.bar(x, accuracies, width, yerr=errors, capsize=5, color='lightgreen')

# Add labels, title, and custom x-axis tick labels
ax.set_xlabel('Models')
ax.set_ylabel('Accuracy')
ax.set_title(f'General accuracies with Confidence Intervals')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylim(0, 1.05)  # Setting the limit of y-axis to a little above 1 for better visibility

# Display the plot
#plt.show()

############# P _ value
from scipy.stats import norm

# Sample data
x1, n1 = 350, 400
x2, n2 = 366, 400

# Proportions
p1 = x1 / n1
p2 = x2 / n2

# Pooled proportion
p_pool = (x1 + x2) / (n1 + n2)

# Standard error
SE = np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))

# Z-score
z = (p1 - p2) / SE

# P-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z)))

# Print the results
print(f"Proportion 1: {p1}")
print(f"Proportion 2: {p2}")
print(f"Z-score: {z}")
print(f"P-value: {p_value}")
```

```python
import seaborn as sns
data = {
    "": ["GPT 3.5", "GPT 4o", "LLama 70B", "Gemini", "Claude 3 Sonnet"],
    "Commonsense": [69, 99, 66, 92, 95],
    "Justice": [70, 93, 76, 62, 90],
    "Virtue and ethics": [73, 84, 86, 85, 87],
    "Deontological": [56, 90, 70, 70, 78]
}

df = pd.DataFrame(data)
df.set_index("", inplace=True)
df

# Plotting the interaction plot
plt.figure(figsize=(12, 6))

# Loop through each model and plot their scores across different datasets
for model in df.index:
    plt.plot(df.columns, df.loc[model], marker='o', label=model)

plt.title('Interaction Plot of Different Models Across Datasets')
plt.xlabel('Datasets')
plt.ylabel('Scores')
plt.legend(title='Models')
plt.grid(True)
plt.show()
```

Når de forskellige linjer krydser hinanden må der være en sammenhæng mellem dataset og model når det kommer til accuracy. Hvis linjerne havde været paralelle, havde det blot været enten datasættet / modellerne der havde påvirket accuracy. 

Vi ser derfor at der er interaction mellem dataset og model. 

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Reshape the DataFrame for two-way ANOVA
df_melted = df.reset_index().melt(id_vars=[""], var_name="Dataset", value_name="Score")
df_melted.columns = ['Model', 'Dataset', 'Score']

# Fit the two-way ANOVA model
model = ols('Score ~ C(Model) + C(Dataset)', data=df_melted).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Print the ANOVA table
print(anova_table)
df
```
